```python
import pandas as pd
```

```python
artist_name = pd.read_csv("unique_artists.txt")
artist_style = pd.read_csv("artist_term.csv")
artist_data = pd.read_csv("artist_data.txt", sep="\t",error_bad_lines=False, names=["id", "name"])
```

```python
artist = []
for name in artist_name["name"]:
    temp = name.split("<SEP>")
    artist.append([temp[0],temp[-1]])
```

```python
artist_name_parsed = pd.DataFrame(artist, columns=['artist_id', 'name'])
```

```python
artist_name_parsed
```

```python
df = artist_style.join(artist_name_parsed, lsuffix='artist_id', rsuffix='artist_id', how="inner")
```

```python
df.drop('artist_idartist_id', axis=1, inplace=True)
```

```python
df.to_csv("artist_style.txt", index=False, header=False, sep = "\t")
```
